# Numeric and Categorical Variable Visualizations

```python
plt.figure(figsize=(15,10))
sns.scatterplot(data=df, x="Age", y="Final Weight", hue='Income')
```

```python
plt.figure(figsize=(15,10))
sns.histplot(df, x="Final Weight", y="Education Level", hue="Sex")
```

```python
plt.figure(figsize=(15,10))
sns.histplot(data=df, x="Final Weight", hue="Income", multiple="stack", log_scale=True)
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(data=df, x="Final Weight", y="Education Level", hue="Sex", orient='h',
               split=True)
sns.despine(left=True)
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x="Race", y="Final Weight",
            hue="Income", palette=["m", "g"],
            data=df)
sns.despine(offset=10, trim=True)
```

```python
sns.pairplot(df, hue="Income")
```
