# Comparing the categorical variables Sex, Race, and Education Level with Income based on Final Weight


### Education, Sex, and Income

In the bar graph below "Income based on Education Level" it appears that there is a pattern of making equal to or under 50,000 US dollars per year, if you have an education level of highschool or lower. In contrast, individuals with an education level of some college or greater are more likely to have an income of over 50,000 US dollars per year.

```python
plt.figure(figsize=(15,8))
ax_Ed_Sex = sns.barplot(x="Education Level", y='Final Weight', hue='Income', data=df,).set_title('Figure 1: Income based on Education Level')
plt.show()
```

Since there are some trends when considering an individuals anual income based on their education, lets see which sex is more likely to have particular levels of education. Let's consider the violin plot and bar plot below.

```python
plt.figure(figsize=(10,10))
sns.violinplot(data=df, x="Final Weight", y="Education Level", hue="Sex", orient='h',
               split=True).set_title('Income based on Education level and Sex')
sns.despine(left=True)
```

```python
plt.figure(figsize=(15,8))
ax_Ed_Sex = sns.barplot(x="Education Level", y='Final Weight', hue='Sex', data=df,).set_title('Education Level Organized by Sex')
plt.show()
```

The barplot and violin plot above show that overall males are more likely than females to have higher education levels based on a final weighted score. Although, it's important to note that the differences are small for males and females with a highschool diploma and masters. It's also important to note that females are more likely to have a doctorate than males. Although there are some differences in education among males and females, let's refer back to our research question and and focus on whether or not there differences are in regards to their income. See figure bellow.

```python
plt.figure(figsize=(8,4))
ax_Ed_Sex = sns.barplot(x="Income", y='Final Weight', hue='Sex', data=df,).set_title('Income Oraganized by Sex')
plt.show()
```

In the figure above males are more likely to make both under or equal to 50K and over 50K. Based on the graph above as well as the graph showing  "Education Level Organized by Sex" it would appear that the slight differences in education level among males and females could correlate with males being more likely to make 50K, while also explaining the larger proportion of males with lower education levels being more likely to make under 50k. 

### Relationships between Race, Education, and Income
Now let's look as some trends in regards to race and income. Consider the bar plot below.

```python
plt.figure(figsize=(15,10))
ax_Ed_Sex = sns.barplot(x="Income", y='Final Weight', hue='Race', data=df,).set_title('Income based on Race by Final Weight')
plt.show()
```

The bar plot above is categorized by income level based on race. As you can see Black Americans are most likely to make under 50,000 dollars, but also most likely to make over 50 000 dollars based on their final weighted score. It is important to note that there was a very small sample of Black Americans (312 individuals) to participate in this census, whereas there were over 20000 White americans; this suggests that the actual number of Black Americans with an income of over 50K is very small. 

Consider the graph below that shows the number of individuals and their income based on race.

```python
ax_income = sns.countplot(y="Race", hue='Income', data=df,).set_title('Income based on Race by Number')
plt.legend(loc='lower right')
plt.show()
```

Now let's look at the next graph and consider whether or not education level could also be a contributing factor when considering income based on race. 

```python
plt.figure(figsize=(15,10))
ax_Ed_Sex = sns.barplot(x="Education Level", y='Final Weight', hue='Race', data=df,).set_title('Income based on Race')
plt.show()
```

In the graph above, you can see that Black Americans are more likely than other races to have the education level in all categories except a doctorate, prof-school, elementary, or preschool. White americans are most likely to have their doctorate, followed by Asian Pacific Islanders. American Indian Eskimo's are most likely to have there education in a prof-school and least likely to have their doctorate.
